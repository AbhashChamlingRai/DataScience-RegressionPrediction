```python
def get_descriptive_information_of_categoric_columns(input_a_categoric_feature):
    '''
    Gives information (prints) about given categorical features such as:
        1. All unique values and number of those unique values with no outlier present in their price distribution.
        2. All unique values and number of those unique values with outlier present in their price distribution.
        3. Total number of unique values in that categorical feature.
        4. Most expensive unique value name in that categorical feature.
        5. Cheapest unique value name in that categorical feature.
    '''
    no_outliers_list = []
    outliers_present_list = []

    highest_name, highest_val = '', 0
    lowest_name, lowest_val = '', 0

    element_name = input_a_categoric_feature
    for i in df[element_name].unique():
        inner_df = df[(df[element_name]== i)]

        #Figuring out the upper bound of the price of each brand so as to compare the prices of every brand and fgure out which has the highest and lowest price
        # Upper bound = Q3 + 1.5 * (Q3-Q1)
        upper_bound = inner_df["price_in_aud"].quantile(0.75) + 1.5 * (inner_df["price_in_aud"].quantile(0.75) - inner_df["price_in_aud"].quantile(0.25))

        #To figure out any outliers we also need lower bound and upper which we already have
        # lower bound = Q1 – 1.5 * (Q3-Q1)
        lower_bound = inner_df["price_in_aud"].quantile(0.25) - 1.5 * (inner_df["price_in_aud"].quantile(0.75) - inner_df["price_in_aud"].quantile(0.25))

        if i == df[element_name].unique()[0]:
            highest_name, highest_val = i, upper_bound
            lowest_name, lowest_val = i, lower_bound
        else:
            if highest_val>upper_bound:
                pass
            elif highest_val<upper_bound:
                highest_name, highest_val = i, upper_bound
            elif lowest_val<lower_bound:
                pass
            elif lowest_val>lower_bound:
                lowest_name, lowest_val = i, lower_bound

        if inner_df.loc[inner_df["price_in_aud"]>upper_bound].shape[0] != 0 or inner_df.loc[inner_df["price_in_aud"]<lower_bound].shape[0] != 0:
            outliers_present_list.append(i)
        elif inner_df.loc[inner_df["price_in_aud"]>upper_bound].shape[0] == 0 or inner_df.loc[inner_df["price_in_aud"]<lower_bound].shape[0] == 0:
            no_outliers_list.append(i)

    print(f"In the dataset, there are {len(no_outliers_list)} unique values in '{element_name}' feature whose price distribution have no outliers. They are :")
    for num, i in enumerate(no_outliers_list):
        print(f"    {num+1}. {i}")
    print()

    print(f"In the dataset, there are {len(outliers_present_list)} unique values in '{element_name}' feature whose price distribution have outliers present in them. They are :")
    for num, i in enumerate(outliers_present_list):
        print(f"    {num+1}. {i}")
    print()
    print(f"In the dataset, there are {len(df[element_name].unique())} unique values.")
    print(f"In the dataset, the unique value with the highest price amount is '{highest_name}'")
    print(f"In the dataset, the unique value with the cheapest price amount is '{lowest_name}'")
```
